```python
import urllib.request
import pandas as pd
import boto3
```

```python
def extract_data(url, filename):
  try:
    urllib.request.urlretrieve(url, filename)
  except Exception as e:
    print(e)
```

```python
extract_data(url='https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/c9509ab4-6f6d-4b97-979a-0cf2a10c922b/download/tmphrybkxuh.csv', filename='data/data_2015.csv')
extract_data(url='https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/b7ea6b1b-3ca4-4c5b-9713-6dc1db52379a/download/tmpzxzxeqfb.csv', filename='data/data_2016.csv')
extract_data(url='https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/30022137-709d-465e-baae-ca155b51927d/download/tmpzccn8u4q.csv', filename='data/data_2017.csv')
extract_data(url='https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/2be28d90-3a90-4af1-a3f6-f28c1e25880a/download/tmp7602cia8.csv', filename='data/data_2018.csv')
extract_data(url='https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/ea2e4696-4a2d-429c-9807-d02eb92e0222/download/tmpcje3ep_w.csv', filename='data/data_2019.csv')
extract_data(url='https://data.boston.gov/dataset/8048697b-ad64-4bfc-b090-ee00169f2323/resource/6ff6a6fd-3141-4440-a880-6f60a37fe789/download/tmpcv_10m2s.csv', filename='data/data_2020.csv')
```

```python
arquivos = [
    'data/data_2015.csv',
    'data/data_2016.csv',
    'data/data_2017.csv',
    'data/data_2018.csv',
    'data/data_2019.csv',
    'data/data_2020.csv'
]
```

```python
dfs = {}
```

```python
for arquivo in arquivos:
  ano = arquivo.split('_')[-1].split('.')[0]
  dfs[ano] = pd.read_csv(arquivo)
```

# Parquet

```python
from io import BytesIO
```

```python
for ano, df in dfs.items():
  parquet_buffer = BytesIO()
  df.to_parquet(parquet_buffer)

  s3.put_object(
      Bucket = 'datalake-comune',
      Key = f'bronze/dados_{ano}.parquet',
      Body = parquet_buffer.getvalue()
  )
```

```python
response = s3.list_objects(Bucket = 'datalake-comune')
```

```python
keys = [obj["Key"] for obj in response["Contents"]]
print(keys)
```
